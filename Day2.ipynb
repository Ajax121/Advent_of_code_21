```python
df = pd.read_csv('input.txt',sep=' ',header=None,names=['dire','value'])
df['value'] = pd.to_numeric(df.value)
df
```

```python
f = df[df['dire']=='forward']['value'].sum()
d = df[df['dire']=='down']['value'].sum()
u = df[df['dire']=='up']['value'].sum()
f*(d-u)
```

```python
depth =0
DEPTHS = []

for indx,row in df.iterrows():
    if row['dire'] == 'down':
        depth +=  row['value']
    elif row['dire'] == 'up':
        depth -= row['value']
    DEPTHS.append(depth)

df['depth'] = DEPTHS
df
```
